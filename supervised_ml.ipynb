## CLASSIFICATION synthetic data

make_forge() is a simple, fixed, 2-class synthetic dataset used for demo in mglearn which internally uses  specific, deprecated version of the much more complex make_blobs(); a general-purpose, highly customizable function within scikit-learn for generating synthetic clustering datasets.

```python
import mglearn
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y = make_blobs()

mglearn.discrete_scatter(X[:,0], X[:,1], y)
plt.legend(["Class 0", "Class 1"], loc=4)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
print(f"X.shape: {X.shape} {y}")
```

```python
import mglearn
from mglearn.datasets import make_forge
import matplotlib.pyplot as plt

X, y = make_forge()

#Plot dataset....26 Data points with 2 features

mglearn.discrete_scatter(X[:,0], X[:,1], y)
plt.legend(["Class 0", "Class 1"], loc=4)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
print(f"X.shape: {X.shape} {y}")
```

## REGRESSION synthetic data

```python
from mglearn.datasets import make_wave
import matplotlib.pyplot as plt

# ...single input feature with continuous target variable or response

X, y = make_wave(n_samples=40)

plt.plot(X, y, 'o')
plt.ylim(-3, 3)
plt.xlabel("Feature")
plt.ylabel("Target")
plt.show()
```

## REAL-WORLD DATA

- Wisconsin Breast Cancer Dataset (Classification - Check if tumors are benign or malignant)
- Boston Housing Dataset (Regression - predict average value of homes in the 70's using crime rate, highway a11y etc)......*Deprecated*

```python
from sklearn.datasets import load_breast_cancer
import numpy as np
cancer = load_breast_cancer()

print(f"Cancer keys: \n {cancer.keys()}")
print("Shape of cancer data: {}".format(cancer.data.shape))
print("Sample counts per class:\n{}".format(
    {
        n: v for n, v in zip(cancer.target_names, np.bincount(cancer.target))
    }
))
print("Feature names:\n{}".format(cancer.feature_names))
```
